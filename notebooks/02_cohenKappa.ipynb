Check inter-rater agreement using Cohen's kappa on the manually-labelled sample.

```python
from sklearn.metrics import cohen_kappa_score
from pathlib import Path
import pandas as pd

manual_sample = pd.read_csv(
    Path("../data/outputs/manual_labeling_sample_150.csv"))
kappa = cohen_kappa_score(
    manual_sample["reliance_label_label"], manual_sample["manual_label"])
print(f"Cohen's Kappa: {kappa}")
```
