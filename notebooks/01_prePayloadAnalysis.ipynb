# Pre-payload Analysis
Checking the dataset before prepping for batch processing.

First, inspect some lines from the dataset:

```python
import json
import random
import os

filePath = os.path.abspath('../../StudyChat/data.jsonl')


def getRandomJSONLLines(filePath, size=50):
    reservoir = []

    with open(filePath, 'r', encoding='utf-8') as file:
        for i, line in enumerate(file):
            if not line.strip():
                continue

            if i < size:
                reservoir.append(json.loads(line))
            else:
                s = random.randint(0, i)
                if s < size:
                    reservoir[s] = json.loads(line)

    for item in reservoir:
        print(json.dumps(item, indent=2))


getRandomJSONLLines(filePath)
```

Find the 10 longest unique `chatIDs` by `chatTotalInteractionCount`.

```python
import heapq
import json
import os

filePath = os.path.abspath('../../StudyChat/data.jsonl')


def streamUniqueLines(filePath):
    seenIDs = set()
    with open(filePath, "r", encoding="utf-8") as f:
        for line in f:
            try:
                data = json.loads(line)
                lineID = data.get("chatId")

                if lineID is None or lineID in seenIDs:
                    continue

                seenIDs.add(lineID)
                yield data, line
            except json.JSONDecodeError:
                continue


longestUniqueEntries = heapq.nlargest(
    10, streamUniqueLines(filePath), key=lambda item: item[0].get("chatTotalInteractionCount", 0)
)

top10 = [item[0] for item in longestUniqueEntries]
print(json.dumps(top10, indent=2))
```
